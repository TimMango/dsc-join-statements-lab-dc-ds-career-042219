## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
import sqlite3
import pandas as pd
con = sqlite3.connect("data.sqlite")
c= con.cursor()
df = pd.read_sql_query('''SELECT * from employees;''', con)
df.head()
```

## Display the names of all the employees in Boston.
Hint: join the employees and customers tables.

```python
df = pd.read_sql_query('''SELECT offices.city, employees.firstName, employees.lastName
               FROM offices
                   LEFT JOIN employees
                   ON offices.officeCode
               WHERE city = 'Boston';''', con)
df
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
df2 = pd.read_sql_query('''SELECT employees.employeeNumber, employees.officeCode, offices.officeCode
               FROM employees
               JOIN offices
               ON offices.OfficeCode;''', con)
df2
```
